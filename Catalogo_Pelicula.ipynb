```python
# Programa de Peliculas.

class Pelicula:
  def __init__(self, titulo, duracion, lanzamiento):
    self.titulo = titulo
    self.duracion = duracion
    self.lanzamiento = lanzamiento
    print(f'Se ha creado la pelicula: {self.titulo}')

  def __str__(self):
    return f'{self.titulo} ({self.lanzamiento})'

"""
Se crea la parte catalogo.
"""

class Catalogo:
  peliculas= [] # diccionario vacio

  def __init__(self, peliculas=[]):
    Catalogo.peliculas = peliculas

  def agregar(self, p): # p es el objeto Pelicula
    Catalogo.peliculas.append(p)

  def mostrar(self):
    for p in Catalogo.peliculas:
      print(p)

    # Instaciamos
peli_1 = Pelicula('Rescatando al Soladado Ryan', 130, 1998)

  # Catalogo que contiene una pelicula

c = Catalogo([peli_1])

c.mostrar()                #Problemas con la identaciones...

c.agregar(Pelicula('Enemigo a las puertas', 230, 2001))

c.mostrar()
```
